```python
import requests
import pandas as pd
```

```python
url_alumnas = "https://beta.adalab.es/curso-intensivo-fullstack-recursos/apis/alumnos-v1/alumnos.json"
```

```python
res = requests.get(url_alumnas)
```

```python
res.status_code
```

```python
res.content
```

```python
res.json()
```

```python
datos_alumnas = res.json()
```

```python
len(datos_alumnas)
```

```python
type(datos_alumnas)
```

```python
datos_alumnas["results"]
```

```python
len(datos_alumnas["results"])
```

```python
datos_alumnas["results"][3]["nombre"]
```

```python
for datos in datos_alumnas["results"]:
    print(datos)
```

```python
for datos in datos_alumnas["results"]:
    print(datos["num_expediente"], datos["nombre"],datos["apellidos"])
```

```python
datos_alumnas_pd = {"num_expediente":[],"nombre":[], "apellidos":[]}
```

```python
for datos in datos_alumnas["results"]:
    datos_alumnas_pd["num_expediente"].append(datos["num_expediente"])
    datos_alumnas_pd["nombre"].append(datos["nombre"])
    datos_alumnas_pd["apellidos"].append(datos["apellidos"])
```

```python
datos_alumnas_pd
```

```python
df = pd.DataFrame(datos_alumnas_pd)

df.head()
```
